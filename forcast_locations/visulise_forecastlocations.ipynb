```python
import folium
import json

c_lon = -3.39
c_lat = 57.00
```

```python
map_uk = folium.Map(location=[c_lat,c_lon], width=1500, height=1000, zoom_start=14)
folium.CircleMarker([c_lat,c_lon], 
                    radius=5,
                    popup='Gazetteer location of Braemar',
                    colour='#3186ec', 
                    fill_color='#3186ec',
                    fill_opacity=1).add_to(map_uk)
folium.CircleMarker([c_lat,c_lon], 
                    radius=100,
                    popup='Gazetteer location of Braemar',
                    colour='#3186ec').add_to(map_uk)
folium.CircleMarker([c_lat,c_lon], 
                    radius=200,
                    popup='Gazetteer location of Braemar',
                    colour='#3186ec').add_to(map_uk)
folium.CircleMarker([c_lat,c_lon], 
                    radius=300,
                    popup='Gazetteer location of Braemar',
                    colour='#3186ec',
                    fill_color='#3186ec',
                    fill_opacity=0.1).add_to(map_uk)
```

```python
f = open('forecast_locations.json')
forecast_locations = json.load(f)
for i in forecast_locations:
    if i['domestic'] == True:
        lat = i['position']['lat']
        lon = i['position']['lon']
        name = i['name']
        
        if (lat < c_lat + 0.2 and lat > c_lat - 0.2 and lon < c_lon + 0.2 and lon > c_lon - 0.2):
            #print(f"c_lat: {c_lat}")
            #print(f"lat: {lat}")
            folium.Marker([lat,lon], popup=f"Forcast Location: {name}").add_to(map_uk)
map_uk
```
